```python
import os
import random as rd
import numpy as np
```

```python
# La longitud vendrá dada por la palabra que busquemos
print("Introduzca palabra a buscar")
#Sentencia_target = input() # Cromosoma
Sentencia_target = 'PUTO GORDO MUERETE'
NPOB = 150
NGEN = 1000
Pm = 0.9878787888787878
NRES = 100
NSAMPLE = 200
```

```python
# La sentencia sólo puede tener valores mayúscula y espacio. 
sentencia = Sentencia_target.upper()
ltar= len(sentencia) #longitud sentencia target
genes = ["A","B","C","D","E","F","G","H","I","J","K","L","M","N","Ñ","O","P","Q","R","S","T","U","V","W","X","Y","Z"," "]
print('Palabra objetivo: '+ sentencia)
```

```python
def aleatorio(b, t):
    random_data = os.urandom(8)

    #Coge la sucesión de carácteres criptográficos obtenidos y los transformamos a un int, actua como semilla
    seed = int.from_bytes(random_data, byteorder="big") 
    rd.seed(seed)
    
    return (rd.randint(b,t)/t)
```

```python
def nuevo_individuo(longitud_sentencia):
    individuo= []
    for x in range(0,(ltar)):
        ale = rd.randint(0,len(genes)-1) # Número aleatorio que será el caracter(gen) del nuevo individuo.
        individuo.append(genes[ale])
    return (individuo)
```

```python
def nueva_poblacion(longitud_sentencia, NPOB):
    poblacion = []
    for i in range(NPOB):
        poblacion.append(nuevo_individuo(longitud_sentencia))
    
    return(poblacion)
```

```python
def fitness_poblacion(poblacion, sentencia):
    coincidencias = []
    fitness_pob = []
    for individuo in poblacion:
        ncoin=0

        for x in range(0,len(individuo)):
            if individuo[x]== sentencia[x]:
                ncoin+=1
        coincidencias.append(ncoin)   
        F= np.exp(ncoin-ltar)-np.exp(-ltar)
        fitness_pob.append(F)
    return(fitness_pob, coincidencias)
```

```python
def replicacion(poblacion):
    #Calculo de la probabilidad de seleccion
    ps = []
    for x in range(0,NPOB):
        seleccion = fitness_pob[x]/np.exp(max(coincidencias)-ltar)-np.exp(-ltar)
        ps.append(seleccion)
        montecarlo = aleatorio(0,100000)
    
    # Individuo inicial
    
    i = rd.randint(0,NPOB-1)
    # Seleccionar ultimo individuo
    if (i==0):
        fin = NPOB-1 
    else:
        fin = i-1
        
    # Pasar por cada individuo
    while (not i == fin):
        
        if (ps[i]>=montecarlo):
            #print('REPRODUCCION')
            
            aleat = rd.randint(0,len(poblacion)-1)
            #print(poblacion[aleat])
            #print(poblacion[x])
            poblacion[aleat] = poblacion[i].copy()
            
        #Mutación del padre
            for y in range(0,ltar):
                #Probabilidad Montecarlo
                montecarlo=aleatorio(0,1000)
                
                #Pm es la probabilidad de MUTAR
                if(Pm>=montecarlo):
                    #Mutación
                    #aleat porque es si muta su hijo
                    poblacion[i][y] = genes[rd.randint(0,len(genes)-1)] #Gen aleatoria
                
        if (i == NPOB-1):
            i=0
        else:
            i +=1
    return(poblacion)
```
